```python
def angle_between(row,key):
    x1, y1 = [row[angles[key]['x1']['x']],row[angles[key]['x1']['y']]]
    x2, y2 = [row[angles[key]['x2']['x']],row[angles[key]['x2']['y']]]
    x3, y3 = [row[angles[key]['x3']['x']],row[angles[key]['x3']['y']]]
    deg1 = (360 + degrees(atan2(x1 - x2, y1 - y2))) % 360 -180
    deg2 = (360 + degrees(atan2(x3 - x2, y3 - y2))) % 360  -180
    return deg2 - deg1 if deg1 <= deg2 else 360 - (deg1 - deg2)
```

```python
data=[]
angle_data={}
for index,row in pd_all_data.iterrows():

    for key in angles:
        if(not(key in angle_data)):
                angle_data[key]=np.array([])
        angle_data[key]=np.append(angle_data[key],angle_between(row,key)) 
    if(not('rep_id' in angle_data)):
        angle_data['rep_id']=np.array([])
        angle_data['label']=np.array([])
    angle_data["rep_id"]=np.append(angle_data['rep_id'],row['rep_id'])
    angle_data["label"]=np.append(angle_data["label"],row['label'])

pd_angle_data = pd.DataFrame.from_dict(angle_data)

# pd_all_data=pd_all_data.transpose()

pd_angle_data 
```
