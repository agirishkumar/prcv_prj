```python
def create_json(path,images_keys,annotations_keys):
    f = open(path)
    data = json.load(f)
    k={}

    for img in data['images']:
        k[img['id']]={}
        for key in images_keys:
            k[img['id']][key]=img[key]
    for annot in data['annotations']:
        for key in annotations_keys:
            k[annot['image_id']][key]=annot[key]
            
    return k
    
```

```python
def check_folder_create(finalPath,final):
      k=final.split('/')
      path=finalPath
      for i in range(0,len(k)):
          path=join(path,k[i])
          isExist = exists(path)
          if not isExist:
            # Create a new directory because it does not exist
            mkdir(finalPath)
```
